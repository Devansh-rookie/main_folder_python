II. Write a Python program to combine each line from the first file with the corresponding line in the second file. 

```python
f1 = open("file1.txt","r")
f2 = open("file2.txt","r")

a= f1.readlines()
b= f2.readlines()
l_a = len(a)
l_b = len(b)
c=""
if(l_a>l_b):
    for i in range(l_b):
        c+=a[i]+b[i]
    for i in range(l_a-l_b):
        c+=(a[l_b+i])
else:
    for i in range(l_a):
        c+=a[i]+b[i]
    for i in range(l_b-l_a):
        c+=(a[l_b+i])
for i in c:
    print(c)
```

V.Write a Python program to create a file where all letters of the English alphabet are listed by specified number of letters on each line.

```python
f = open("alphabets_by_number.txt","w")
ch = 1
while(True):
    ch = input("Input 'q' to exit and any digit to continue in the loop: ")
    if ch=='q':
        break
    elif not ch.isnumeric():
        print("Give valid input.")
    else:
        for i in range(int(ch)):
            to_write = str(chr(i+65))+" "
            f.write(to_write)
        f.write("\n")
f.close()
```
