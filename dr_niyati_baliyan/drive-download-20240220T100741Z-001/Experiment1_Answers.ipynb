```python
def check_prime(n):
  flag =0
  if(n>2):
    for i in range(2,n):
      if(n%i==0):
        flag =1
        return 0
    if(flag==0):
      return 1
  elif(n==2):
    return 1
n= eval(input("What is n: "))
sum=0
for i in range(0,n):
  if(check_prime(i)):
    sum+=i

print("Sum: ",sum)
```
