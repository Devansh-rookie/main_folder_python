```python
def check_leap(year):
  if year%4==0:
    if year%400:
      return 1
    elif year%100:
      return 0
    else:
      return 1
  else:
    return 0
months= ["January","February","March","April","May","June","July","August","September","October","November","December"]
days_months={"January": 31,"February":28,"March":31,"April":30,"May":31,"June":30,"July":31,"August":31,"September":30,"October":31,"November":30,"December":31}
month,year = eval(input("Enter comma separated month and year: "))
if(check_leap(year)):# it will run if the function returns 1 that is that the year is leap
  if month==2:
    print(f"{months[month-1]} {year} has {days_months[months[month-1]]+1} days")
  else:
    print(f"{months[month-1]} {year} has {days_months[months[month-1]]} days")
else:
  print(f"{months[month-1]} {year} has {days_months[months[month-1]]} days")
```

Experiment 3, Question 9

```python
def change_to_num(a): # or this entire thing can be used inside the for loop also
  if a=='A':
    return 10
  elif a=='B':
    return 11
  elif a=='C':
    return 12
  elif a=='D':
    return 13
  elif a=='E':
    return 14
  elif a=='F':
    return 15
  else:
    return int(a)
hex = input("Hex Character: ")
length = len(hex)
deci =0
for i in range(length):
  deci+=(16**i)*change_to_num(hex[-i-1])

print(hex," in decimal is ",deci)
# int(string, base_of_numbers)
```
