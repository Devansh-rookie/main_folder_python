Write a Python program to create two strings from a given string. Create the first string using those
characters which occurs only once and create the second string which consists of multi-time occurring
characters in the said string.

```python
a= input("What is the string: ")
single=""
multi=""
for i in range(len(a)):
  c=1
  for j in range(len(a)):
    if a[i]==a[j] and i!=j:# if we remove the condition of i!=j then c=0 has to initialized as it would automatically consider that c+=1 there when i==j
      c+=1
  if c==1:
    single+=a[i]
  else: # here put elif and put the condition of the current iteration of the strings element should not be present in the multi string as its autometically considered in the single string
    multi+=a[i]
print("Single time occuring characters strings: ",single)
print("Multiple time occuring characters strings: ",multi)
```

Question 5:

Write a Python program to find the index of a given string at which a given substring starts. If the
substring is not found in the given string return 'Not found'.

```python
a= input("What is the main string: ")
b= input("What is the sub string: ")
flag = False
for i in range(len(a)):
  if a[i]==b[0]:
    for j in range(len(b)):
      if (a[i+j]==b[j]):
        flag = True
      else:
        flag = False
        break
    if flag == True:
      print("The string starts at: ",i)
      break
if flag == False:
  print("Not Found")
```

Question 6:

Find the number of Upper case and Lower case characters in a string

```python
a= input("What is the string: ")
c_up=0
c_low=0
for i in a:
  if i.isupper():
    c_up+=1
  elif i.islower():
    c_low+=1

print(f"Uppercase characters are: {c_up}\nLowercase character are: {c_low}")
```

Another Method to do question 6

Question 7:

Check for palindrome

```python
a= input("What is the string: ")
flag = True
for i in range(int(len(a)/2)):
  if (a[i]!=a[len(a)-i-1]):
    print(f"The string is not a palindrome: {a}")
    flag = False
    break
if flag ==True:
  print(f"The string is a palindrome: {a}")
```
